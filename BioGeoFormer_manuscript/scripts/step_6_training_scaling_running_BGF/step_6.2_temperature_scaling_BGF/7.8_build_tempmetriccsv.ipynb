```python
import os
import re
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Path to your temperature_scaling directory
base_dir = "/content/drive/MyDrive/cycformer_run/cycformer_sep2/cycformer/temperature_scaling"

# Regex patterns to extract values
lambda_pattern = re.compile(r"Lambda:\s*(\d+)")
before_pattern = re.compile(r"Before Temp - NLL:\s*([\d.]+), ECE:\s*([\d.]+)")
after_pattern = re.compile(r"After Temp - NLL:\s*([\d.]+), ECE:\s*([\d.]+)")

results = []

# Loop through cycle directories (cyc_20 ... cyc_90)
for cyc_dir in sorted(os.listdir(base_dir)):
    cyc_path = os.path.join(base_dir, cyc_dir)
    if not os.path.isdir(cyc_path) or not cyc_dir.startswith("cyc_"):
        continue

    # Loop through lambda directories
    for lambda_dir in sorted(os.listdir(cyc_path)):
        lambda_path = os.path.join(cyc_path, lambda_dir)
        if not os.path.isdir(lambda_path) or not lambda_dir.startswith("lambda_"):
            continue

        # Find metrics text files
        for file in os.listdir(lambda_path):
            if file.endswith(".txt"):
                file_path = os.path.join(lambda_path, file)
                with open(file_path, "r") as f:
                    content = f.read()

                # Extract metrics
                lambda_match = lambda_pattern.search(content)
                before_match = before_pattern.search(content)
                after_match = after_pattern.search(content)

                if lambda_match and before_match and after_match:
                    results.append({
                        "cycle": cyc_dir,
                        "lambda": int(lambda_match.group(1)),
                        "before_nll": float(before_match.group(1)),
                        "before_ece": float(before_match.group(2)),
                        "after_nll": float(after_match.group(1)),
                        "after_ece": float(after_match.group(2)),
                        "file": file
                    })

# Convert to DataFrame
df = pd.DataFrame(results)

# Sort for readability
df = df.sort_values(by=["cycle", "lambda"]).reset_index(drop=True)

# Show results
import caas_jupyter_tools
caas_jupyter_tools.display_dataframe_to_user("Temperature Scaling Metrics", df)
```
